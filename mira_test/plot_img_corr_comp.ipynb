```python
%matplotlib qt
import matplotlib.pyplot as plt
import numpy as np

fnames = ["20260506_124941-1ms-RAW12-240",
            "20260506_134737-CDS-BOTH_1ms-RAW12-240"]

img_data_stack = np.load(f"./meas/{fnames[0]}.npy")
mean_img = np.mean(img_data_stack, axis=0, keepdims=False)

plt.imshow(mean_img, vmin=150, vmax=420)
plt.axis('off')  # Hide axes for cleaner image display
plt.colorbar(location='bottom', shrink=0.6)
plt.title("RAW12 No Corrections")
plt.savefig(f"./meas/RAW12_No-Corrections.png", dpi=300)
plt.show()
```
